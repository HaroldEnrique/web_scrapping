```python
import urllib.request
import json
import os
import pandas as pd
import requests
```

```python
def get_image(url_image, folder_to_save):
    #urllib.request.urlretrieve(url_image,folder_to_save)
    r = requests.get(url_image,timeout=15)

    with open(folder_to_save, 'wb') as f:
        f.write(r.content)
```

```python
files_names = []
for file in os.listdir("./"):
    if file.endswith(".csv"):
        files_names.append(os.path.join("",file))
files_names
```

```python
for file in files_names:
    name = file.split('.')[0]
    folder_name = name + '_images'
    path_folder = os.path.join(os.getcwd(),folder_name)
    os.mkdir(path_folder)
    print(path_folder)
    df = pd.read_csv('./'+file)
    df = df.fillna('')
    
    lista_img =df[['id_pro','img_url']].values.tolist()
    for img in lista_img:
        if img[1]:
            extension = img[1].split('.')[-1]
            name_img = str(img[0])+'.'+str(extension)
            path_to_save = os.path.join(path_folder,name_img)
            try:
                get_image(img[1],path_to_save)
                print('Success', img[1])
            except Exception as e:
                print('Error > ',e,' URL: ',img[1], 'Folder',folder_name)
                continue
    print('DONE >> ', path_folder)
```
